* 线性判别分析（$\mathrm{LDA}$）
<br /><br />
    * 给定二分类数据集
    $$
    D = D_{+} \cup D_{-}
    $$
    * 设正例样本和负例样本服从正态分布
    $$
    \begin{gather*}
    \mathbf{X}_{+} \sim N(\mu_{+}, \mathbf{\Sigma_{+}}) \\ \\
    \mathbf{X}_{-} \sim N(\mu_{-}, \mathbf{\Sigma_{-}})
    \end{gather*}
    $$
    * 将所有样本投影到直线（一维空间）上实现数据的降维
    $$
    \begin{gather*}
    \tilde{x} = w \cdot \mathbf{x} \\ \\
    \mathcal{E}\tilde{x} = w \cdot \mu \\ \\
    \mathcal{D}\tilde{x} = w^\mathrm{T} \mathbf{\Sigma} w
    \end{gather*}
    $$
    * 为了最大程度地保留数据特征 &rArr; 最大化数据的可分程度
    $$
    \begin{gather*}
    增大类间距离    \Rightarrow [w \cdot (\mu_{+} - \mu_{-})]^2 = w^\mathrm{T} (\mu_{+} - \mu_{-}) (\mu_{+} - \mu_{-})^\mathrm{T} w \\ \\
    减小类内分散程度 \Rightarrow \sigma^{2}_{+} + \sigma^{2}_{-} = w^\mathrm{T} (\mathbf{\Sigma}_{+} + \mathbf{\Sigma}_{-}) w
    \end{gather*}
    $$
    * 依据上式定义类间、类内散度矩阵
    $$
    \begin{gather*}
    \mathbb{S}_{b} = (\mu_{+} - \mu_{-}) (\mu_{+} - \mu_{-})^\mathrm{T} \\ \\
    \mathbb{S}_{w} = \mathbf{\Sigma}_{+} + \mathbf{\Sigma}_{-}
    \end{gather*}
    $$
    * 为了同时达到增大类间距离和减小类内分散程度的目的，定义广义瑞利商
    $$
    J = \frac{w^\mathrm{T} \mathbb{S}_{b} w}{w^\mathrm{T} \mathbb{S}_{w} w}
    $$
    * 线性判别分析
    $$
    \hat{w} = arg \max_{w} J = arg \max{\frac{w^\mathrm{T} \mathbb{S}_{b} w}{w^\mathrm{T} \mathbb{S}_{w} w}}
    $$
    * 由于问题的重点在于确定$w$的方向，为简化问题限制$w$使得
    $$
    w^\mathrm{T} \mathbb{S}_{w} w = 1
    $$
    * 问题转换为
    $$
    \begin{gather*}
    \hat{w} = arg \min_{w} -w^\mathrm{T} \mathbb{S}_{b} w \\ \\
    s.t.\quad w^\mathrm{T} \mathbb{S}_{w} w = 1
    \end{gather*}
    $$
    * 利用拉格朗日乘子法
    $$
    \begin{gather*}
    L(w,\ \lambda) = -w^\mathrm{T} \mathbb{S}_{b} w + \lambda (w^\mathrm{T} \mathbb{S}_{w} w - 1) \\ \\
    \frac{\partial L}{\partial w} = -2\mathbb{S}_{b} w + 2\lambda \mathbb{S}_{w} w = 0
    \end{gather*}
    $$
    * 其中
    $$
    \mathbb{S}_{b} w = (\mu_{+} - \mu_{-}) (\mu_{+} - \mu_{-})^\mathrm{T} w = \gamma(w) (\mu_{+} - \mu_{-})
    $$
    * 代入方程可得
    $$
    \hat{w} = \frac{\gamma(\hat{w})}{\lambda} \mathbb{S}^{-1}_{w} (\mu_{+} - \mu_{-}) \Rightarrow \mathbb{S}^{-1}_{w} (\mu_{+} - \mu_{-})
    $$

* 主成分分析（$\mathrm{PCA}$）
<br /><br />
    * 将$n$个$d$维的样本映射到$d'$个$d$维单位向量张成的线性空间
    $$
    \begin{gather*}
    \mathbf{x}_{k} \longrightarrow \mathbf{m} + a_{k1} \mathbf{e}_{1} + \cdots + a_{kd'} \mathbf{e}_{d'} \\ \\
    s.t.\quad ||\mathbf{e}_{i} || = 1
    \end{gather*}
    $$
    * 为了方便讨论，取该线性空间中的一组正交单位向量作为基向量
    $$
    \begin{gather*}
    \mathbf{x}_{k} \longrightarrow \mathbf{m} + a_{k1} \mathbf{e}_{1} + \cdots + a_{kd'} \mathbf{e}_{d'} \\ \\
    s.t.\quad \mathbf{e}_{i}^{\mathrm{T}} \mathbf{e}_{j} = \mathbb{I}(i = j)
    \end{gather*}
    $$
    * 将映射形式改写为
    $$
    \begin{gather*}
    \mathbf{x}_{k} \longrightarrow \mathbf{m} + \mathbf{E} \mathbf{a}_{k} \\ \\
    s.t.\quad \mathbf{E}^{\mathrm{T}} \mathbf{E} = \mathbf{I}
    \end{gather*}
    $$
    * 为了尽可能减小降维造成的损失，可以最小化平方误差损失函数
    $$
    \begin{gather*}
    \min_{\mathbf{a},\ \mathbf{E},\ \mathbf{m}} J(\mathbf{a},\ \mathbf{E},\ \mathbf{m}) = 
    \sum_{k = 1}^{n} || \left( \mathbf{m} + \mathbf{E} \mathbf{a}_{k} \right) 
    - \mathbf{x}_{k} ||^{2} \\ \\
    s.t.\quad \mathbf{E}^{\mathrm{T}} \mathbf{E} = \mathbf{I}
    \end{gather*}
    $$
    * 暂时只考虑向量的模为1的约束，即
    $$
    \mathbf{\Lambda} = diag(\lambda_{1},\ \cdots,\ \lambda_{d'})
    $$
    * 构造拉格朗日函数
    $$
    \begin{align*}
    \mathcal{L}(\mathbf{a},\ \mathbf{m},\ \mathbf{E},\ \mathbf{\Lambda}) &= J(\mathbf{a},\ \mathbf{E},\ \mathbf{m}) 
    + \mathrm{tr}[\mathbf{\Lambda} (\mathbf{E}^{\mathrm{T}} \mathbf{E} - \mathbf{I})] \\ \\
    &= \sum_{k = 1}^{n} || \left( \mathbf{m} + \mathbf{E} \mathbf{a}_{k} \right) 
    - \mathbf{x}_{k} ||^{2} + \mathrm{tr}[\mathbf{\Lambda} (\mathbf{E}^{\mathrm{T}} \mathbf{E} - \mathbf{I})]
    \end{align*}
    $$
    * 重写原优化问题
    $$
    \begin{gather*}
    \min_{\mathbf{\Lambda}} \min_{\mathbf{E}} \min_{\mathbf{a},\ \mathbf{m}} 
    \mathcal{L}(\mathbf{a},\ \mathbf{m},\ \mathbf{E},\ \mathbf{\Lambda}) \\ \\
    s.t.\quad \mathbf{E}^{\mathrm{T}} \mathbf{E} = \mathbf{I}
    \end{gather*}
    $$
    * 内层的$\min_{\mathbf{a},\ \mathbf{m}} \mathcal{L}$需要满足方程
    $$
    \begin{gather*}
    \frac{\partial \mathcal{L}}{\partial \mathbf{a}_{k}} = 2\mathbf{E}^{\mathrm{T}} 
    [ \mathbf{E} \mathbf{a}_{k} - (\mathbf{x}_{k} - \mathbf{m}) ] = 0 \\ \\
    \frac{\partial \mathcal{L}}{\partial \mathbf{m}} = 
    2\sum_{k = 1}^{n} [ \mathbf{E} \mathbf{a}_{k} - (\mathbf{x}_{k} - \mathbf{m}) ] = 0 \\ \\
    s.t.\quad \mathbf{E}^{\mathrm{T}} \mathbf{E} = \mathbf{I}
    \end{gather*}
    $$
    * 解得
    $$
    \mathbf{a}_{k} = \mathbf{E}^{\mathrm{T}} (\mathbf{x}_{k} - \mathbf{m})
    \qquad
    \mathbf{m} = \frac{1}{n} \sum_{k = 1}^{n} \mathbf{x}_{k}
    $$
    * 代入原拉格朗日函数
    $$
    \mathcal{L}(\mathbf{E},\ \mathbf{\Lambda}) = -\mathrm{tr}[\mathbf{E}^{\mathrm{T}} \mathbb{S}_{t} \mathbf{E}] + 
    \mathrm{tr}[\mathbb{S}_{t}] + \mathrm{tr}[\mathbf{\Lambda} (\mathbf{E}^{\mathrm{T}} \mathbf{E} - \mathbf{I})]
    \Rightarrow -\mathrm{tr}[\mathbf{E}^{\mathrm{T}} \mathbb{S}_{t} \mathbf{E}]+ 
    \mathrm{tr}[\mathbf{\Lambda} (\mathbf{E}^{\mathrm{T}} \mathbf{E} - \mathbf{I})]
    $$
    * 其中，$\mathbb{S}_{t}$代表全局散度矩阵
    $$
    \mathbb{S}_{t} = \sum_{k = 1}^{n} (\mathbf{x}_{k} - \mathbf{m})(\mathbf{x}_{k} - \mathbf{m})^{\mathrm{T}}
    $$
    * 中间层的$\min_{\mathbf{E}} \mathcal{L}$需要满足方程
    $$
    \frac{\partial \mathcal{L}}{\partial \mathbf{E}} = -2 \mathbb{S}_{t} \mathbf{E} + 
    \mathbf{E} (\mathbf{\Lambda} + \mathbf{\Lambda}^{\mathrm{T}}) = 0
    $$
    * 解得
    $$
    \mathbb{S}_{t} \mathbf{E} = diag(\lambda_{1},\ \cdots,\ \lambda_{d'}) \mathbf{E}
    $$
    * 上式限定了$\mathbf{E}$的列向量为全局散度矩阵的$d'$个特征向量，$\lambda$为对应的特征值
    $$
    \mathbb{S}_{t} \mathbf{e}_{i} = \lambda_{i} \mathbf{e}_{i}
    $$
    * 与约束条件$\mathbf{E}^{\mathrm{T}} \mathbf{E} = \mathbf{I}$一起代入原拉格朗日函数
    $$
    \mathcal{L}(\mathbf{\Lambda}) = -\mathrm{tr}[\mathbf{\Lambda}] = -\sum_{i = 1}^{d'} \lambda_{i}
    $$
    * 为了最小化拉格朗日函数$\mathcal{L}$，将$\lambda$取做$\mathbb{S}_{t}$的最大的$d'$个特征值
    <br /><br />
    * 相应地，$\mathbf{E}$的列向量为对应的$d'$个特征向量，考虑到$\mathbf{S}_{t}$为实对称矩阵，特征向量同时又相互正交，满足了剩余的约束条件
