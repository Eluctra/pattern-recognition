* 线性判别分析（$\mathrm{LDA}$）
<br /><br />
    * 给定二分类数据集
    $$
    D = D_{+} \cup D_{-}
    $$
    * 设正例样本和负例样本服从正态分布
    $$
    \begin{gather*}
    \boldsymbol{X}_{+} \sim N(\mu_{+}, \boldsymbol{\Sigma_{+}}) \\ \\
    \boldsymbol{X}_{-} \sim N(\mu_{-}, \boldsymbol{\Sigma_{-}})
    \end{gather*}
    $$
    * 将所有样本投影到直线（一维空间）上实现数据的降维
    $$
    \begin{gather*}
    \tilde{x} = w \cdot \boldsymbol{x} \\ \\
    \mathcal{E}\tilde{x} = w \cdot \mu \\ \\
    \mathcal{D}\tilde{x} = w^\mathrm{T} \boldsymbol{\Sigma} w
    \end{gather*}
    $$
    * 为了最大程度地保留数据特征 &rArr; 最大化数据的可分程度
    $$
    \begin{gather*}
    增大类间距离    \Rightarrow [w \cdot (\mu_{+} - \mu_{-})]^2 = w^\mathrm{T} (\mu_{+} - \mu_{-}) (\mu_{+} - \mu_{-})^\mathrm{T} w \\ \\
    减小类内分散程度 \Rightarrow \sigma^{2}_{+} + \sigma^{2}_{-} = w^\mathrm{T} (\boldsymbol{\Sigma}_{+} + \boldsymbol{\Sigma}_{-}) w
    \end{gather*}
    $$
    * 依据上式定义类间、类内散度矩阵
    $$
    \begin{gather*}
    \mathbb{S}_{b} = (\mu_{+} - \mu_{-}) (\mu_{+} - \mu_{-})^\mathrm{T} \\ \\
    \mathbb{S}_{w} = \boldsymbol{\Sigma}_{+} + \boldsymbol{\Sigma}_{-}
    \end{gather*}
    $$
    * 为了同时达到增大类间距离和减小类内分散程度的目的，定义广义瑞利商
    $$
    J = \frac{w^\mathrm{T} \mathbb{S}_{b} w}{w^\mathrm{T} \mathbb{S}_{w} w}
    $$
    * 线性判别分析
    $$
    \hat{w} = arg \max_{w} J = arg \max{\frac{w^\mathrm{T} \mathbb{S}_{b} w}{w^\mathrm{T} \mathbb{S}_{w} w}}
    $$
    * 由于问题的重点在于确定$w$的方向，为简化问题限制$w$使得
    $$
    w^\mathrm{T} \mathbb{S}_{w} w = 1
    $$
    * 问题转换为
    $$
    \begin{gather*}
    \hat{w} = arg \min_{w} -w^\mathrm{T} \mathbb{S}_{b} w \\ \\
    s.t.\quad w^\mathrm{T} \mathbb{S}_{w} w = 1
    \end{gather*}
    $$
    * 利用拉格朗日乘子法
    $$
    \begin{gather*}
    L(w,\ \lambda) = -w^\mathrm{T} \mathbb{S}_{b} w + \lambda (w^\mathrm{T} \mathbb{S}_{w} w - 1) \\ \\
    \frac{\partial L}{\partial w} = -2\mathbb{S}_{b} w + 2\lambda \mathbb{S}_{w} w = 0
    \end{gather*}
    $$
    * 其中
    $$
    \mathbb{S}_{b} w = (\mu_{+} - \mu_{-}) (\mu_{+} - \mu_{-})^\mathrm{T} w = \gamma(w) (\mu_{+} - \mu_{-})
    $$
    * 代入方程可得
    $$
    \hat{w} = \frac{\gamma(\hat{w})}{\lambda} \mathbb{S}^{-1}_{w} (\mu_{+} - \mu_{-}) \Rightarrow \mathbb{S}^{-1}_{w} (\mu_{+} - \mu_{-})
    $$

* 主成分分析（$\mathrm{PCA}$）
<br /><br />
    * 将$n$个$d$维的样本映射到$d'$个$d$维单位向量张成的线性空间
    $$
    \begin{gather*}
    \boldsymbol{x}_{k} \longrightarrow \boldsymbol{m} + a_{k1} \boldsymbol{e}_{1} + \cdots + a_{kd'} \boldsymbol{e}_{d'} \\ \\
    s.t.\quad ||\boldsymbol{e}_{i} || = 1
    \end{gather*}
    $$
    * 为了方便讨论，取该线性空间中的一组正交单位向量作为基向量
    $$
    \begin{gather*}
    \boldsymbol{x}_{k} \longrightarrow \boldsymbol{m} + a_{k1} \boldsymbol{e}_{1} + \cdots + a_{kd'} \boldsymbol{e}_{d'} \\ \\
    s.t.\quad \boldsymbol{e}_{i}^{\mathrm{T}} \boldsymbol{e}_{j} = \mathbb{I}(i = j)
    \end{gather*}
    $$
    * 将映射形式改写为
    $$
    \begin{gather*}
    \boldsymbol{x}_{k} \longrightarrow \boldsymbol{m} + \boldsymbol{E} \boldsymbol{a}_{k} \\ \\
    s.t.\quad \boldsymbol{E}^{\mathrm{T}} \boldsymbol{E} = \boldsymbol{I}
    \end{gather*}
    $$
    * 为了尽可能减小降维造成的损失，可以最小化平方误差损失函数
    $$
    \begin{gather*}
    \min_{\boldsymbol{a},\ \boldsymbol{E},\ \boldsymbol{m}} J(\boldsymbol{a},\ \boldsymbol{E},\ \boldsymbol{m}) = 
    \sum_{k = 1}^{n} || \left( \boldsymbol{m} + \boldsymbol{E} \boldsymbol{a}_{k} \right) 
    - \boldsymbol{x}_{k} ||^{2} \\ \\
    s.t.\quad \boldsymbol{E}^{\mathrm{T}} \boldsymbol{E} = \boldsymbol{I}
    \end{gather*}
    $$
    * 暂时只考虑向量的模为1的约束，即
    $$
    \boldsymbol{\Lambda} = diag(\lambda_{1},\ \cdots,\ \lambda_{d'})
    $$
    * 构造拉格朗日函数
    $$
    \begin{align*}
    \mathcal{L}(\boldsymbol{a},\ \boldsymbol{m},\ \boldsymbol{E},\ \boldsymbol{\Lambda}) &= J(\boldsymbol{a},\ \boldsymbol{E},\ \boldsymbol{m}) 
    + \mathrm{tr}[\boldsymbol{\Lambda} (\boldsymbol{E}^{\mathrm{T}} \boldsymbol{E} - \boldsymbol{I})] \\ \\
    &= \sum_{k = 1}^{n} || \left( \boldsymbol{m} + \boldsymbol{E} \boldsymbol{a}_{k} \right) 
    - \boldsymbol{x}_{k} ||^{2} + \mathrm{tr}[\boldsymbol{\Lambda} (\boldsymbol{E}^{\mathrm{T}} \boldsymbol{E} - \boldsymbol{I})]
    \end{align*}
    $$
    * 重写原优化问题
    $$
    \begin{gather*}
    \min_{\boldsymbol{\Lambda}} \min_{\boldsymbol{E}} \min_{\boldsymbol{a},\ \boldsymbol{m}} 
    \mathcal{L}(\boldsymbol{a},\ \boldsymbol{m},\ \boldsymbol{E},\ \boldsymbol{\Lambda}) \\ \\
    s.t.\quad \boldsymbol{E}^{\mathrm{T}} \boldsymbol{E} = \boldsymbol{I}
    \end{gather*}
    $$
    * 内层的$\min_{\boldsymbol{a},\ \boldsymbol{m}} \mathcal{L}$需要满足方程
    $$
    \begin{gather*}
    \frac{\partial \mathcal{L}}{\partial \boldsymbol{a}_{k}} = 2\boldsymbol{E}^{\mathrm{T}} 
    [ \boldsymbol{E} \boldsymbol{a}_{k} - (\boldsymbol{x}_{k} - \boldsymbol{m}) ] = 0 \\ \\
    \frac{\partial \mathcal{L}}{\partial \boldsymbol{m}} = 
    2\sum_{k = 1}^{n} [ \boldsymbol{E} \boldsymbol{a}_{k} - (\boldsymbol{x}_{k} - \boldsymbol{m}) ] = 0 \\ \\
    s.t.\quad \boldsymbol{E}^{\mathrm{T}} \boldsymbol{E} = \boldsymbol{I}
    \end{gather*}
    $$
    * 解得
    $$
    \boldsymbol{a}_{k} = \boldsymbol{E}^{\mathrm{T}} (\boldsymbol{x}_{k} - \boldsymbol{m})
    \qquad
    \boldsymbol{m} = \frac{1}{n} \sum_{k = 1}^{n} \boldsymbol{x}_{k}
    $$
    * 代入原拉格朗日函数
    $$
    \mathcal{L}(\boldsymbol{E},\ \boldsymbol{\Lambda}) = -\mathrm{tr}[\boldsymbol{E}^{\mathrm{T}} \mathbb{S}_{t} \boldsymbol{E}] + 
    \mathrm{tr}[\mathbb{S}_{t}] + \mathrm{tr}[\boldsymbol{\Lambda} (\boldsymbol{E}^{\mathrm{T}} \boldsymbol{E} - \boldsymbol{I})]
    \Rightarrow -\mathrm{tr}[\boldsymbol{E}^{\mathrm{T}} \mathbb{S}_{t} \boldsymbol{E}]+ 
    \mathrm{tr}[\boldsymbol{\Lambda} (\boldsymbol{E}^{\mathrm{T}} \boldsymbol{E} - \boldsymbol{I})]
    $$
    * 其中，$\mathbb{S}_{t}$代表全局散度矩阵
    $$
    \mathbb{S}_{t} = \sum_{k = 1}^{n} (\boldsymbol{x}_{k} - \boldsymbol{m})(\boldsymbol{x}_{k} - \boldsymbol{m})^{\mathrm{T}}
    $$
    * 中间层的$\min_{\boldsymbol{E}} \mathcal{L}$需要满足方程
    $$
    \frac{\partial \mathcal{L}}{\partial \boldsymbol{E}} = -2 \mathbb{S}_{t} \boldsymbol{E} + 
    \boldsymbol{E} (\boldsymbol{\Lambda} + \boldsymbol{\Lambda}^{\mathrm{T}}) = 0
    $$
    * 解得
    $$
    \mathbb{S}_{t} \boldsymbol{E} = diag(\lambda_{1},\ \cdots,\ \lambda_{d'}) \boldsymbol{E}
    $$
    * 上式限定了$\boldsymbol{E}$的列向量为全局散度矩阵的$d'$个特征向量，$\lambda$为对应的特征值
    $$
    \mathbb{S}_{t} \boldsymbol{e}_{i} = \lambda_{i} \boldsymbol{e}_{i}
    $$
    * 与约束条件$\boldsymbol{E}^{\mathrm{T}} \boldsymbol{E} = \boldsymbol{I}$一起代入原拉格朗日函数
    $$
    \mathcal{L}(\boldsymbol{\Lambda}) = -\mathrm{tr}[\boldsymbol{\Lambda}] = -\sum_{i = 1}^{d'} \lambda_{i}
    $$
    * 为了最小化拉格朗日函数$\mathcal{L}$，将$\lambda$取做$\mathbb{S}_{t}$的最大的$d'$个特征值
    <br /><br />
    * 相应地，$\boldsymbol{E}$的列向量为对应的$d'$个特征向量，考虑到$\boldsymbol{S}_{t}$为实对称矩阵，特征向量同时又相互正交，满足了剩余的约束条件
